```python
import mesa as ms
from Models import *
from Agents import *
import pandas as pd
import matplotlib.pyplot as plt
```

```python
num_iterations = 10
num_of_steps   = 100
```

```python
params = {"nCars"   :   range(2, 4) }

results = ms.batch_run(
    RoomModel,
    parameters=params,
    iterations=num_iterations,
    max_steps=num_of_steps,
    number_processes=1,
    data_collection_period=1,
    display_progress=True
)
```

```python
results_df = pd.DataFrame(results)
print(results_df.keys())
results_df
```
